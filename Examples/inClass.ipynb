Load our dataset and prepare mini batches

```python
import torch 
import torchvision
import matplotlib
import numpy as np

normalize_transforms = torchvision.transforms.Compose(
    [torchvision.transforms.ToTensor(),
     torchvision.transforms.Normalize(mean=(0.5,0.5,0.5), std=(0.5,0.5,0.5))]
)

trainData = torchvision.datasets.CIFAR100(root='./CIFAR100/train', train=True, transform=normalize_transforms, download=True)
testData = torchvision.datasets.CIFAR100(root='./CIFAR100/test', train=False, transform=normalize_transforms, download=True)

print("\nLOADED DATASETS")

batchSize = 128
trainDataLoader = torch.utils.data.DataLoader(
    trainData, batch_size = batchSize
)

testDataLoader = torch.utils.data.DataLoader(
    testData, batch_size = batchSize
)
```

Build our Model! Yay

```python
class CNN(torch.nn.model):
    def __init__(self):
        super().__init__()
        self.model = torch.nn.Sequential(
            torch.nn.Conv2d(in_channels = 3, out_channels = 5, kernel_size = 3, padding = 1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size = 2),
            torch.nn.Conv2d(in_channels = 5, out_channels = 8, kernel_size = 3, padding = 1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2),
            torch.nn.Flatten(),
            torch.nn.Linear(8*8*8, 300),
            torch.nn.Linear(300, 100)
        )
    
    def forward(self, x):
        return self.model(x)
```
